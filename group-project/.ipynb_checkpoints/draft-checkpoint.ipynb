```python
import numpy as py
import pandas as pd
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('netflix_titles.csv')
df.head()
```

```python
df.drop(df.loc[df['type']!='Movie'].index, inplace=True)
df
```

```python
df.dropna(inplace= True)
df.head()
```

```python
month = df['date_added'].str.split(pat = ' [1-9]',n = 1, expand = True)
df['date_added'] = month[0]
df
```

```python
genresplit = df['listed_in'].str.split(pat = ",", expand = True)
genresplit['date_added'] = df['date_added']
a = genresplit.groupby(['date_added'], axis = 0).count()
```

```python
fig ,axes= plt.subplots(figsize=(20,20))
axes.bar(genresplit(['date_added'], genresplit[0]))
```
