```python
import numpy as py
import pandas as pd
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('netflix_titles.csv')
df.head()
```

```python
df.drop(df.loc[df['type']!='Movie'].index, inplace=True)
df
```

```python
df.dropna(inplace= True)
df.head()
```

```python
month = df['date_added'].str.split(pat = ' [1-9]',n = 1, expand = True)
df['date_added'] = month[0]
df
```

```python
genresplit = df['listed_in'].str.split(pat = ",", expand = True)
genresplit['date_added'] = df['date_added']
for name, group in genresplit.groupby('date_added'):
    print(name)
    print(group)
    
a = dict(list(genresplit.groupby('date_added')))
a['January']
```

```python
plt.figure(figsize=(13,8))
sns.boxplot(y='release_year', x='rating', data=df, palette='BuGn')
sns.despine(left=True)
plt.ylim(1990,2020)
```
