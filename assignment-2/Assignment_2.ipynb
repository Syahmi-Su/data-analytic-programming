```python
#Numpy Task
import numpy as np
data = np.loadtxt('Attendance_RAW.txt')
x = data.reshape((14,15))
attendance_arr = np.transpose(x)
print (x)
```

```python
#Numpy Task3
def avg_day():
    result = []
    day = np.arange(14)
    total = attendance_arr.sum(axis= 0)
    avg_att = total / 15
    avg_day = np.around(avg_att, decimals = 2)
    for i in range(len(day)):
        result.append(["Day " + str(day[i]+1) , avg_day[i]])
    print(result)



avg_day()    
```

```python
#Pandas Task1 & Task2
import pandas as pd
data = pd.read_csv('emp_detail.csv')
data.head()
```

```python
#Task3
attendance = pd.DataFrame(attendance_arr)
emp_attendance = pd.concat([data, attendance], axis = 1)
emp_attendance
```

```python
#Task4 & Task5
total = emp_attendance.sum(axis = 1)
emp_attendance['total'] = total
rate = total / 14 * 100
emp_attendance['rate'] = rate
emp_attendance.round(2)
```

```python
#Task6
status = np.where(rate > 90, 'Good', (np.where(rate < 60, 'Bad', 'Moderate')))
```

```python
emp_attendance['status'] = status
emp_attendance
```

```python
def department():
    finance_arr = emp_attendance[emp_attendance['Department'] == 'FINANCE']
    staff_arr = emp_attendance[emp_attendance['Department'] == 'STAFFING']
    avg_finance = finance_arr['total'].mean()
    avg_staff = staff_arr['total'].mean()
    average_data = {'Deparment': ["FINANCE", 'STAFFING'],
                    'average': [avg_finance,avg_staff]}
    avg_arr = pd.DataFrame(average_data)

    return avg_arr

department()
```
